```python
#2
df_Dummies = pd.get_dummies(df_NoMissing, columns = ['rep_education'])
df_Dummies
#Creates dummy variables.
```

```python
y = df_Dummies["Def_ind"]
x = df_Dummies.loc[:, df_Dummies.columns != "Def_ind"]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, )
#Seperate into training and test sets. Def_ind is the dependent variable. 
```

```python
#3
model = LogisticRegression()
model.fit(X_train, y_train)
model.coef_
#Fitting the model.
```

```python
print(model.intercept_, model.coef_,model.score(X_test, y_test))
#This gives me the intercept (-9.98748593e-07), the coefficients, and the R squared (0.90333) meaning 90% of the variance in defaults can be explained by the model.
```

```python
#4
y_pred = model.predict(X_test)
confusion = confusion_matrix(y_test,y_pred)
print(confusion)
#This produces the confusion matrix. The output gives us 2964 true positives (model predicted the account defaulted and it did), 25 true negatives (model predicted teh account did not default and it did not default), 14 false positives (model predicted the accoint defaulted but it did not), and 328 false negatives (model predicted the account did not default but it did default). 
```

```python
#5
y_score = model.predict_proba(X_test)[:, 1]
```

```python
falsePositiveRate, truePositiveRate, thresholds = roc_curve(y_test, y_score)
roc_auc = auc(falsePositiveRate, truePositiveRate)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(falsePositiveRate, truePositiveRate, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc='lower right')
plt.show()
#Plotting the ROC/AUC curve.This tells us the performance of our model. The closer the ROC curve is to the high, left corner on the y axis (true positive rate), the better our model is.
```

```python
#6
coefs = model.coef_
variables = X_train.columns
print(coefs, variables)
#The variables that affect Def_ind the most are tot_amount_currently_past_due (coef = 2.64271008e-04) and num_card_inq_24_month (coef= 6.27586729e-05)
#This makes sense as the total amount past due and the number of inquiries would affect if someone defaults or not.
```

```python
#2
y2 = df2["price_range"]
x2 = df2.loc[:, df2.columns != "price_range"]
X_train2, X_test2, y_train2, y_test2 = train_test_split(x2, y2, test_size=0.3, )
#Seperating the data set into training and test sets.
```

```python
#3
lda = LinearDiscriminantAnalysis()
model2 = lda.fit(X_train2, y_train2)
model2.coef_
#Fitting the LDA model and getting coefficients.
```

```python
print(model2.intercept_, model2.coef_,model2.score(X_test2, y_test2))
#This gives me the intercepts, coefficients, and the R squared (0.945) meaning 94.5% of the variation in price range can be explained by the model.
```

```python
#4
y_pred2 = lda.predict(X_test2)

accuracy = accuracy_score(y_test2, y_pred2)
precision = precision_score(y_test2, y_pred2, average='weighted')
recall = recall_score(y_test2, y_pred2, average='weighted')
#Getting the accuracy, precision, and recall.
```

```python
print(accuracy)
print(precision)
print(recall)
```

```python
#The accuracy is high (0.948333333) meaning the model makes a correct prediction between postive and negative predictions 94.8% of the time.
#The precision is high (0.9505019) meaning that 95% of the time the model predicts a positive class, it is positive.
#The recall is also high (0.948333333) meaning the model rarely misses positive cases. 
```

```python
#5

confusion2 = confusion_matrix(y_test2,y_pred2)
print(confusion2)
#produces a confusion matrix for the model.
#The diagonal line represents the true positives for each class. So for price range=0, there were 135 true positives, price range=1 there were 146 true positives, for price range =2 there were 139 true positives, and for price range=3 there were 147.
#This tells me that the model does a good job of predicting the price range. 
```
